4. Now let us construct a Liouvillian numerically. Consider the Lindblad equation for a qubit in a thermal bath $$ \frac{{\rm d}}{{\rm d}t} \hat{\rho} = \frac{1}{{\rm i}\hbar} [ \hat{H},\hat{\rho}] + \gamma (1+\bar{n})\mathcal{D}[\hat{\sigma}_-]\hat{\rho} + \gamma \bar{n} \mathcal{D}[\hat{\sigma}_+]\hat{\rho} = \mathcal{L}\hat{\rho},$$ where $\hat{H} = \hbar\omega_0 \hat{\sigma}_z/2$, $\bar{n} = ({\rm e}^{\beta\hbar\omega_0}-1)^{-1}$ and $\beta = 1/k_BT$ is the inverse temperature. Take the parameters to be $\omega_0=1$, $\gamma = 0.1\omega_0$ and $k_BT = 0.5\hbar\omega_0$.

(a) The Pauli operator basis is defined by $\hat{e}_j = \hat{\sigma}_j/\sqrt{2}$, where $\{\hat{\sigma}_0,\hat{\sigma}_1,\hat{\sigma}_2,\hat{\sigma}_3\}= \{\hat{\mathbb{1}},\hat{\sigma}_x,\hat{\sigma}_y,\hat{\sigma}_z\}$. Verify numerically that this forms an orthonormal basis with respect to the inner product $\langle \hat{e}_i,\hat{e}_j\rangle = {\rm tr}(\hat{e}_i^\dagger \hat{e}_j)$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg

sigma = [[],[],[],[]]

sigma[0] = (np.eye(2))/np.sqrt(2)
sigma[1] = np.array([[0,1],[1,0]])/np.sqrt(2)
sigma[2] = np.array([[0,-1j],[1j,0]])/np.sqrt(2)
sigma[3] = np.array([[1,0],[0,-1]])/np.sqrt(2)

# With our Basis formed we can trivially check if the defined inner products of our basis
# returns a kronecker delta.

for i in range(4):
    for j in range(4):
        print(f'<e_{i},e_{j}> = ', np.trace(sigma[i].conj().T@sigma[j]))

print('We can see that these Basis Elements form an orthonormal basis with the given definition of the inner product')
```

(b) Construct and print the Liouvillian operator $\hat{\mathcal{L}}$ as a NumPy array in the Pauli basis.

```python
sigma_plus = 0.5*(np.array([[0,1],[1,0]])+1j*np.array([[0,-1j],[1j,0]]))
sigma_minus = 0.5*(np.array([[0,1],[1,0]])-1j*np.array([[0,-1j],[1j,0]])) 


def Commutator(A,B):
  return A@B - B@A

def AntiCommutator(A,B):
  return A@B + B@A

Ham = np.array([[1,0],[0,-1]])/2

def n_bar(b):
  return (np.exp(b)-1)**-1

beta = 2

def D(L,rho):
  return L@rho@L.conj().T - 0.5*AntiCommutator(L.conj().T@L,rho)

def L(gamma,ei,ej):
  return np.trace(ei.conj().T@(Commutator(Ham,ej)*-1j + gamma*(1+n_bar(beta))*D(sigma_minus,ej) + gamma*n_bar(beta)*D(sigma_plus,ej)))

Lou = np.zeros((4,4),dtype = complex)

for i, si in enumerate(sigma):
  for j, sj in enumerate(sigma):
    Lou[i][j] = L(0.1,si,sj)

print('Louvillian = ',Lou)
```

(c) Find the eigenvalues and right eigenvectors of $\hat{\mathcal{L}}$. Print the numerically obtained eigenvectors in their standard matrix form (i.e. expressed in the computational basis $|0\rangle$ and $|1\rangle$). Verify that the steady state is thermal as expected. Comment on the implications of these results for the time evolution of the qubit.

```python
eigenvalues = np.linalg.eig(Lou)[0]
eigenvectors = [[],[],[],[]]
minpos = abs(eigenvalues).argmin()
steady_state =  np.linalg.eig(Lou)[1][:,minpos].flatten()

for i in range(4):
  eigenvectors[i] = np.linalg.eig(Lou)[1][:,i].flatten()

print('steady state in pauli basis = ' , steady_state.reshape(2,2),'\n')

vec = np.zeros(4)
def Basis_Change(basis,vector):
  for i in range(4):
    vec[i] = vector@basis[i].reshape(4,1).flatten()
  return vec

for k in range(4):
  eigenvectors[k] = (Basis_Change(sigma,eigenvectors[k])).reshape(2,2)
  

print('eigenvectors in comp basis = ', eigenvectors)


Comp_vec = (Basis_Change(sigma,steady_state)/np.trace(Basis_Change(sigma,steady_state).reshape(2,2))).reshape(2,2)

print('Steady State in Computational Basis = \n', Comp_vec)




                   

print('Thermal State for this configuration = \n',scipy.linalg.expm(-beta*Ham)/np.trace(scipy.linalg.expm(-2*Ham)))

print('We can therefore see that our Steady state is thermal.')
```

```python
Ham = np.array([[1,0],[0,-1]])/2

def ML(H):
  return np.kron(H,np.eye(len(H)))

def MR(H):
  return np.kron(np.eye(len(H)),H.conj().T)

def Super_Ham(H):
  return ML(H) - MR(H)

print ('H = ',Super_Ham(Ham),'\n\n','H_dagger = ' ,Super_Ham(Ham).conj().T)

print('\nWe can see that H = H_dagger and it is therefore Hermitian.')
```
