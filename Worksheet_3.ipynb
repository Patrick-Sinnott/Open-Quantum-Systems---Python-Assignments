4. Consider the *cat state*

```python
def Q(x,y):
    v1 = np.exp(-abs(x-y)**2)
    v2 = np.exp(-abs(-x-y)**2)
    v3 = np.exp(-abs(x-y)**2-2*x*y)
    v4 = np.exp(-abs(-x-y)**2+2*x*y)    
    N_plus = 2*(1+np.exp(-2*abs(x)**2))
    return 1/(np.pi*N_plus)*(v1+v2+v3+v4)

alpha = np.linspace(-5,5,100)
X, Y = np.meshgrid(alpha,alpha)

num = X+1j*Y
print(Q(1+1j,1))
plt.figure()
plt.contourf(X,Y,Q (num,1))
plt.colorbar()

plt.figure()
plt.contourf(X,Y,Q (num,3))
plt.colorbar()
```

```python
def FockKet(d,n):
  Ket = np.zeros(d)
  Ket[n] = 1
  return Ket

def FockBra(d,n):
  return FockKet(d,n).conj().T
  
def annihilation(d):
  a=0
  for n in range (d-1):
    a += np.sqrt(n+1)*np.outer(FockKet(d,n),FockBra(d,n+1))
  return a

def creation(d):
  return annihilation(d).conj().T

def commutator(a,b):
  return np.dot(a,b) - np.dot(b,a)

print (annihilation(5),'\n')

print ('a|4> =',np.dot(annihilation(5),FockKet(5,4)),'\n')
print ('|3>=',FockKet(5,3),'\n')


print('[a,a_dag]=',commutator(annihilation(5),creation(5)))


```

```python
dt = 0.1
def U(dt,H): 
  return linalg.expm(-1j*H*dt)

psi_0 = FockKet(16,0)
psi_t = FockKet(16,0)
q0 = 1

q = (annihilation(16) + creation(16))

def Expt_value(psi,op): 
  return psi.conj().T@op@psi

t_range = np.arange(0, 100, dt)

Traj1 = np.zeros(t_range.size)
Expt_Values1 = np.zeros(t_range.size)

for i in range(t_range.size):
  Traj1[i] = psi_0.conj().T@psi_t
  psi_t  = U(dt, H1(1,t_range[i],0.99,0.1))@psi_t

psi_t = FockKet(16,0)

for i in range(t_range.size):
  Expt_Values1[i] = Expt_value(psi_t,q)
  psi_t  = U(dt, H1(1,t_range[i],0.99,0.1))@psi_t

plt.plot(t_range,Traj1)
plt.title('Dynamics of the Quantum State')
plt.xlabel('$\omega_0$t')
plt.ylabel('< 0 | $\psi$(t)>') 

plt.figure()
plt.plot(t_range,Expt_Values1)
plt.title('Expectation value of position over time')
plt.xlabel('$\omega_0$t')
plt.ylabel('<q>/$q_0$') 
```

As in question 4, consider a two-mode system with annihilation operators $\hat{a}_1$ and $\hat{a}_2$. Convince yourself that the Hamiltonian can be written as a vector-matrix-vector product ([a quadratic form](https://en.wikipedia.org/wiki/Quadratic_form))
$$\hat{H}  = \mathbf{\hat{a}}^\dagger\cdot \mathbf{H} \cdot \mathbf{\hat{a}} = \sum_{j,k} H_{jk} \hat{a}_j^\dagger \hat{a}_k= 
\begin{pmatrix} \hat{a}_1^\dagger & \hat{a}_2^\dagger \end{pmatrix} \cdot \begin{pmatrix}
\hbar\omega_1  & \hbar g/2 \\ \hbar g/2 & \hbar\omega_2 \end{pmatrix}\cdot \begin{pmatrix} \hat{a}_1 \\ \hat{a}_2 \end{pmatrix},$$
where all three expressions above are just different ways of writing the same thing, with
$$ \mathbf{\hat{a}} = \begin{pmatrix} \hat{a}_1 \\ \hat{a}_2 \end{pmatrix},\qquad \mathbf{\hat{a}}^\dagger = \begin{pmatrix} \hat{a}_1^\dagger & \hat{a}_2^\dagger \end{pmatrix}, \qquad \mathbf{H} = \begin{pmatrix}
\hbar\omega_1  & \hbar g/2 \\ \hbar g/2 & \hbar\omega_2 \end{pmatrix},$$
while $H_{jk}$ are the elements of the $2\times 2$ Hermitian matrix $\mathbf{H}$.

(a) Consider the displacement vector $$\mathbf{d} = \begin{pmatrix}\langle \hat{a}_1 \rangle \\
\langle \hat{a}_2\rangle\end{pmatrix},$$
and the correlation matrix $$\mathbf{C} = \begin{pmatrix}\langle\!\langle \hat{a}_1^\dagger \hat{a}_1 \rangle\!\rangle & \langle\!\langle \hat{a}_2^\dagger \hat{a}_1 \rangle\!\rangle \\
\langle\!\langle \hat{a}_1^\dagger \hat{a}_2 \rangle\!\rangle &\langle\!\langle\hat{a}_2^\dagger \hat{a}_2 \rangle\!\rangle
\end{pmatrix},$$
or, equivalently, $d_j = \langle a_j\rangle$ and $C_{jk} = \langle\!\langle \hat{a}_k^\dagger \hat{a}_j\rangle\!\rangle = \langle \hat{a}_k^\dagger \hat{a}_j\rangle - \langle \hat{a}_k^\dagger\rangle\langle\hat{a}_j\rangle$ (note the order of indices $j,k$!). Use the Heisenberg equations to show that
$$ \frac{\rm d}{\rm dt}\mathbf{d} = \frac{1}{{\rm i}\hbar}\mathbf{H}\cdot \mathbf{d},\qquad  \frac{\rm d}{\rm dt} \mathbf{C} = \frac{1}{{\rm i}\hbar}[\mathbf{H},\mathbf{C}].$$ 

(b) Convince yourself that the solutions to the above equations are $$\mathbf{d}(t) = \mathbf{V}(t)\cdot\mathbf{d}(0),\qquad \mathbf{C}(t) = \mathbf{V}(t)\cdot\mathbf{C}(0)\cdot\mathbf{V}^\dagger(t),$$ where $\mathbf{V}(t) = {\rm e}^{-{\rm i}\mathbf{H} t/\hbar}$ is a unitary matrix. Therefore, the dynamics can be simulated using the same iterative method as we have previously used for qubits in Worksheet 1, since $$\mathbf{d}(t +\Delta t) = \mathbf{V}(\Delta t) \cdot \mathbf{d}(t), \qquad \mathbf{C}(t+\Delta t) = \mathbf{V}(\Delta t)\cdot\mathbf{C}(t)\cdot\mathbf{V}^\dagger(\Delta t).$$ Use this method to calculate the dynamics of the system numerically, starting from the product state

```python
def Ham(w1,g,w2):
  return np.array([[w1,g/2],[g/2,w2]])

def V(dt):
  return linalg.expm(-1j*Ham(1,0.3,0.9)*dt)

psi = np.kron(FockKet(5,4),FockKet(1,0))

d = np.array([0,0])
d_dag = np.array([0,0])
C = np.array([[4,0],[0,0]])
dt = 0.1
t_range = np.arange(0,50,dt)


q1_values = np.zeros(t_range.size)
q2_values = np.zeros(t_range.size)
n1_values = np.zeros(t_range.size)
n2_values = np.zeros(t_range.size)


for i in range(t_range.size):
  q1_values[i] = d[0]+d_dag[0]
  q2_values[i] = d[1]+d_dag[1]
  n1_values[i] = C[0,0]
  n2_values[i] = C[1,1] 
  d = V(dt)@d
  d_dag = V(dt)@d_dag
  C = V(dt)@C@V(dt).conj().T

plt.plot(t_range,q1_values,label = '<$q_1$>')
plt.plot(t_range,q2_values,label = '<$q_2$>')
plt.plot(t_range,n1_values,label = '<$n_1$>')
plt.plot(t_range,n2_values,label = '<$n_2$>')
plt.legend()
```

This models a bath comprising a one-dimensional chain of oscillators coupled to the system at one of its ends ($j=1$).
```python
def Ham_0(v):
  return v*np.eye(4)

def Ham_1(W,G,M):
  H = np.zeros((4,4))
  H[1,0] = H[0,1] = np.sqrt(W*G)/2
  for j in range(1,3):
    H[j,j+1]= W/2
    H[j+1,j]= W/2 
  return H

def Ham(v,W,G,M):
  return Ham_0(v) + Ham_1(W,G,M) 

def C(b,Ham):
  return (linalg.expm(b*Ham) - np.eye(4))**-1

b = 0.5

print(C(b,Ham(2,0.5,0.05,4)))
```
